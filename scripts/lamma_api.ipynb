```python
model_name = "TheBloke/Llama-2-13B-chat-GGML"
model_basename = "llama-2-13b-chat.ggmlv3.q5_1.bin"
```

```python
from huggingface_hub import hf_hub_download
from llama_cpp import Llama
```

```python
model_path = hf_hub_download(repo_id=model_name, filename=model_basename)
```

```python
lcpp_llm = None
lcpp_llm = Llama(model_path=model_path,n_threads=2, n_batch=512, n_gpu_layers=32)
```

```python
prompt = "Please tell the disease based on the following symptoms:"
symptoms = "I have been having a lot of problems with my skin. It is constantly peeling and I have developed rashes on my arms. It is really itchy and it is making my life miserable."
prompt_template=f'''SYSTEM: Provide only the name of the most probable disease based on the following symptoms, with no additional text or explanation. Write the disease name on a new line immediately following this prompt.
USER: {prompt}{symptoms}
ASSISTANT:
'''
```

```python
prompt = "Please tell the disease based on the following symptoms:"
symptoms = "I have been having a lot of problems with my skin. It is constantly peeling and I have developed rashes on my arms. It is really itchy and it is making my life miserable."
prompt_template=f'''SYSTEM: Based on the symptoms described below, respond ONLY with the name of the most probable disease. Do not include any explanations, sentences, or additional text. Write your response on a new line immediately following this prompt.
USER: {prompt}{symptoms}
ASSISTANT:
'''
```

```python
response = lcpp_llm(prompt=prompt_template, max_tokens=256, temperature=0.4, top_p=0.9,
                    repeat_penalty=1.5, top_k=100, echo=True)
```

```python
print((response["choices"][0]["text"]).strip().split("\n")[-1])
```
